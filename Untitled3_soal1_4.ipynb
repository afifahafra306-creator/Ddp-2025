```python
def bilangan(angka):
    for i in range(1,angka):
       if i % 2 != 0:
           print(i, end=", ")
bilangan(20)
```

```python
def nilai(n = 0):
    if n <= 60:
        print("tidak lulus")
    elif n > 60 and n <= 100:
        print("lulus")
    else:
        print("tidak diketahui")
nilai(80)
```

```python
def is_genap(n):
    return n % 2 == 0

print(is_genap(2))
```

```python
def celcius_ke_fahrenheit(celcius):
  fahrenheit = (celcius * 9/5) + 32
  return fahrenheit

print(celcius_ke_fahrenheit(0))
print(celcius_ke_fahrenheit(100))
```
